## 크롤링
  - 크롤링은 총 2번 실행
  - 첫번째 크롤링은 목록 페이지로, 번호, 제목, 글쓴이, 조회, 추천, 주소, 내용 가져옴
  - 두번째 크롤링은 클릭해서 들어간 페이지로 등록일과 내용 가져옴

```python
url_list = []

for i in range(fr, to+1):
    ur = 'https://www.inven.co.kr/board/df/3911?p='+str(i)
#     driver.get(ur)
    time.sleep(1)
    response = requests.get(ur)
    html = response.text
#     html = driver.page_source
#     html = requests.get(ur)
    soup = BeautifulSoup(html , 'html.parser')
    lines = soup.find_all('tr')

    for line in lines:
        try:
            number = line.select_one('.num').text.strip()
            print(number)
            if number == '공지':
                continue
            
            url = line.select_one('.subject-link').get('href')
            print(url)

            title = line.select_one('.subject-link').text.replace('\n','').replace('  ','').strip()
            print(title)
            
            nickname = line.select_one('.layerNickName').text
            print(nickname)
            
            view = int(line.select_one('.view').text.replace(',', ''))
            print(view)
            
            reco = int(line.select_one('.reco').text.replace(',', ''))
            print(reco)
            
            df=df.append({'번호': number, '제목': title, '글쓴이': nickname, '조회': view, '추천': reco, '주소': url},
                         ignore_index=True)
            print('*'*100)
        except:
            print('something goes wrong'*10)


url_list
```

```python
df
```

```python
for i in range(0, len(df)):
    ur = df.loc[i, '주소']
    response = requests.get(ur)
    html = response.text
    time.sleep(0.5)

    soup = BeautifulSoup(html , 'html.parser')
    
    content = soup.select_one('div#powerbbsContent').text.strip()
    print(content)
    
    datetime_string = soup.select_one('.articleDate').text.strip()
#     datetime_string = "2021-12-31 13:35:42.657813"
    datetime_format = "%Y-%m-%d %H:%M"
    datetime_result = datetime.strptime(datetime_string, datetime_format)
    print(datetime_result)
    
    compare_date = datetime.strptime('2020-01-09 00:00', datetime_format)
    
    if datetime_result >= compare_date:
        df.loc[i, '등록일'] = datetime_result
        df.loc[i, '내용'] = content
    else:
        df = df.dropna(subset=["등록일"])
        print('droppppp')
        break

    print('*'*100)
    time.sleep(0.5)
```
